# **Lets start our analysis**

# What can we learn about different hosts and areas?

```python
hosts_areas=df.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].count().reset_index()
hosts_areas.sort_values(by='calculated_host_listings_count',ascending=False).head(5)
```



* The most number of bookings are in Manhattan by Sonder(NYC)  




# what can we learn from predictions ?(locations , prices, reviews, etc...)




```python
areas_reviews = df.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
areas_reviews.sort_values(by='number_of_reviews',ascending=False).head(5)
```

```python
area = areas_reviews['neighbourhood_group']
review = areas_reviews['number_of_reviews']

fig = plt.figure(figsize=(10,5))


plt.bar(area , review , color = 'blue', width = 0.5)
plt.xlabel('AREA')
plt.ylabel('REVIEW')
plt.title('AREA V/S NUMBER OF REVIEWS')
plt.show()
```

```python
price_area = df.groupby(['price'])['number_of_reviews'].max().reset_index()
price_area.head()
```

```python
area = price_area['price']
price = price_area['number_of_reviews']

fig = plt.figure(figsize=(10,5))

#plotting a graph

plt.scatter(area , price , color='red')

plt.xlabel('price')
plt.ylabel('Number of reviews')
plt.title('PRICE VS NUMBER OF REVIEWS')
plt.show()
```

From the above analysis we can say that most people prefer to stay in place where price is less.

# Let's check availability of neighbourhood

```python
availability_type=df.groupby(['neighbourhood_group','room_type'])['availability_365'].max().reset_index()
availability_type=availability_type.sort_values(by='availability_365',ascending=False).head(10)
availability_type
```

```python
f,ax = plt.subplots(figsize=(10,8))
ax=sns.boxplot(x='neighbourhood_group',y='availability_365',data=df,palette="bright")
plt.title("NEIGHBOURHOOD GROUP VS  ROOM AVAILABILITY")
plt.xlabel('NEIGHBOURHOOD GROUP')
plt.ylabel('AVAILABILITY_365')
plt.show()
```

  

*   Satern island is available more than 300 days in a year. Hence it is less prefered by people.




# Which hosts are the busiest and why?

```python
busiest_host = df.groupby (['host_name','host_id','room_type'])['number_of_reviews'].max().reset_index()
busiest_host=busiest_host.sort_values(by='number_of_reviews',ascending=False).head(10)
busiest_host
```

```python
name = busiest_host['host_name']
reviews = busiest_host['number_of_reviews']

fig = plt.figure(figsize = (10,5))

#creating bar graph  

plt.bar(name , reviews , color = 'violet' , width =0.5)

plt.xlabel('NAME OF THE HOST')
plt.ylabel('NUMBER OF REVIEWS')
plt.title('BUSIEST HOSTS')
plt.show()
```

From the above analysis we can see that busiest host is Dona

*   From the above analysis we can see that busiest host is Dona. Dona is the most reviewed host with more thean 600 reviews

*   More preferred room type is private room and entire room. 






# Is there any noticeable difference of traffic among different areas and what could be the reason for it?

Here we are checking the number of people visited in different neighbourhood 

*   Here checking the number of people visited in different neighbourhood 
*   Analyse with number of reviews and neighbourhood groups.





```python
most_reviewed = df.groupby(['neighbourhood_group'])['number_of_reviews'].sum().reset_index()
most_reviewed.sort_values(by='number_of_reviews',ascending =False).head(10)
```

```python
traffic_areas = df.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
traffic_areas = traffic_areas.sort_values(by='minimum_nights', ascending=False)
traffic_areas                                    
```

```python
room_type = traffic_areas['room_type']
stay=traffic_areas['minimum_nights']

fig = plt.figure(figsize=(10,5))

#cretting bar graph
plt.bar(room_type,stay, color='orange',width = 0.5)

plt.xlabel('ROOM TYPE')
plt.ylabel('MINIMUM NUMBER OF NIGHTS STAYED')
plt.title('TRAFFIC AREAS')
plt.show()
```

We can see that more  people prefer entire home/apt and then private room located in Manhattan ,Queens and Brooklyn.

# Find more demanded neighbourhood

```python
availability = df.groupby(['neighbourhood','availability_365'])['minimum_nights'].count().reset_index()
availability = availability.sort_values(by='minimum_nights', ascending=False).head(5)
availability
```

```python
neighbourhood = availability['neighbourhood']
mininum_night = availability['minimum_nights']

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(neighbourhood, mininum_night, color ='green',
        width = 0.5)
 
plt.xlabel("neighbourhood")
plt.ylabel("Minimum_night")
plt.title("More Prefered Neighbourhood")
plt.show()
```



*   Williamsburg is more prefered neighbourhood. Because minimum night stayed is highest in Williamburg.
